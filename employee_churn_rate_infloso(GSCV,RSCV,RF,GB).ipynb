```python
# import lib for data cleaning
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load data set as df
df=pd.read_csv('churn.csv')
```

```python
# check for missig values
df.isnull().sum()
```

```python
# information of load data set with data types need to convert objects into flot or int
df.info()
```

```python
# fill missing values using sample for gender column
gender_distribution = df['gender'].value_counts(normalize=True)
missing_indices = df['gender'].isnull()
missing_count = missing_indices.sum()
filled_values = gender_distribution.sample(n=missing_count, replace=True).index
df.loc[missing_indices, 'gender'] = filled_values
```

```python
# fill missing values using sample for education_level column
gender_distribution = df['education_level'].value_counts(normalize=True)
missing_indices = df['education_level'].isnull()
missing_count = missing_indices.sum()
filled_values = gender_distribution.sample(n=missing_count, replace=True).index
df.loc[missing_indices, 'education_level'] = filled_values
```

```python
# fill missing values using sample for experience column
gender_distribution = df['experience'].value_counts(normalize=True)
missing_indices = df['experience'].isnull()
missing_count = missing_indices.sum()
filled_values = gender_distribution.sample(n=missing_count, replace=True).index
df.loc[missing_indices, 'experience'] = filled_values
```

```python
# fill missing values using sample for company_type column
gender_distribution = df['company_type'].value_counts(normalize=True)
missing_indices = df['company_type'].isnull()
missing_count = missing_indices.sum()
filled_values = gender_distribution.sample(n=missing_count, replace=True).index
df.loc[missing_indices, 'company_type'] = filled_values
```

```python
# fill missing values using sample for company_size column
gender_distribution = df['company_size'].value_counts(normalize=True)
missing_indices = df['company_size'].isnull()
missing_count = missing_indices.sum()
filled_values = gender_distribution.sample(n=missing_count, replace=True).index
df.loc[missing_indices, 'company_size'] = filled_values
```

```python
# fill missing values using sample for major_discipline column
gender_distribution = df['major_discipline'].value_counts(normalize=True)
missing_indices = df['major_discipline'].isnull()
missing_count = missing_indices.sum()
filled_values = gender_distribution.sample(n=missing_count, replace=True).index
df.loc[missing_indices, 'major_discipline'] = filled_values
```

```python
# fill missing values using sample for last_new_job column
gender_distribution = df['last_new_job'].value_counts(normalize=True)
missing_indices = df['last_new_job'].isnull()
missing_count = missing_indices.sum()
filled_values = gender_distribution.sample(n=missing_count, replace=True).index
df.loc[missing_indices, 'last_new_job'] = filled_values
```

```python
# Import lable encoder form sklearn for lable encoding  
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
```

```python
# convert data of into desire form for ML model
df['city'] = df['city'].str.replace('city_', '')
df['city'] = df['city'].astype(int)
df['company_size']= df['company_size'].replace({'10/49': '10-49','10000+':'>10000'})
```

```python
# lable encoding
education_level=df['education_level']=labelencoder.fit_transform(df['education_level'])
relevent_experience=df['relevent_experience']=labelencoder.fit_transform(df['relevent_experience'])
gender=df['gender']=labelencoder.fit_transform(df['gender'])
enrolled_university=df['enrolled_university']=labelencoder.fit_transform(df['enrolled_university'])
company_type=df['company_type']=labelencoder.fit_transform(df['company_type'])
company_size=df['company_size']=labelencoder.fit_transform(df['company_size'])
major_discipline=df['major_discipline']=labelencoder.fit_transform(df['major_discipline'])
last_new_job=df['last_new_job']=labelencoder.fit_transform(df['last_new_job'])
experience=df['experience']=labelencoder.fit_transform(df['experience'])
```

```python
# type conversion
df['target'] = df['target'].astype(int)
```

```python
# all data is converted into flot or int
df.info()
```

```python
# 0 missing values 
df.isnull().sum()
```

```python
# correlation matrix for feature selection
correlation_matrix = df.corr()
correlation_matrix
```

```python
# data visualization of correlation matrix
plt.figure(figsize=(10, 8))
heatmap = sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.yticks(rotation=0)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# 'city_development_index' (correlation: 0.341665)
# 'relevent_experience' (correlation: 0.183165)
# 'enrolled_university' (correlation: 0.141363)
# this 3 are highly corelate with target variable 
absolute_correlation_values = correlation_matrix.abs()
most_correlated = (absolute_correlation_values.where(np.triu(np.ones(absolute_correlation_values.shape), k=1).astype(bool)).stack().sort_values(ascending=False))
print("Most Correlated Features:")
print(most_correlated.head(10))
```

```python
# data assigne to x and y var
x=df[['city_development_index','relevent_experience','enrolled_university']]
y=df[['target']]
```

```python
# lib import
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
# train test split
x_train,x_test,y_train,y_test=train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Import ML algo
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.model_selection import RandomizedSearchCV
```

```python
# model assigne 
model=GradientBoostingClassifier()
```

```python
# finding best parameters for algo 
# parameters={'learning_rate':[0.01,0.02,0.03],
#             'subsample':[0.9,0.5,0.2],
#             'n_estimators':[100,500,1000],
#             'max_depth':[4,6,8]
#            }
# grid=GridSearchCV(estimator=model,param_grid=parameters,cv=2,n_jobs=-1)
```

```python
# applying best parameter to algo
parameters={'learning_rate':[0.03],
            'subsample':[0.5],
            'n_estimators':[100],
            'max_depth':[4]
           }
grid=GridSearchCV(estimator=model,param_grid=parameters,cv=2,n_jobs=-1)
grid.fit(x_train,y_train)
```

```python
# grid score is accuracy of model 
grid.best_score_
```

```python
# predict value for given input
x_new = np.array([[0.920,0,2]]) #0.920(city_development_index) 0(relevent_experience),2(enrolled_university) due to lable encoding
prediction = grid.predict(x_new)
print("Prediction: {}".format(prediction))
```
