```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('heart_disease.csv')
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
x_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
reg=LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
sns.heatmap(df)
```

```python
df.isnull().sum()
```

```python
parameter={'penalty':['l1','l2','l3'],'C':[1,2,3,4,5,6,10,20,30,40,50],'max_iter':[100,200,300]}
```

```python
class_reg=GridSearchCV(reg,param_grid=parameter,scoring='accuracy',cv=5)
```

```python
class_reg.fit(x_train,y_train)
```

```python
print(class_reg.best_params_)
```

```python
print(class_reg.best_score_)
```

```python
df.corr()
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pre=class_reg.predict(x_test)
```

```python
score=accuracy_score(y_pre,y_test)
print(score)
```

```python
y_pre
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_pre,y_test))
```
