```python
data=data.dropna()
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
data.head()
```

```python
data['Dependents'] = data['Dependents'].replace('3+', 4)
```

```python
# gender m-1,f-0
# loan_status n-0,y-1
# married y-1,n-0
# education graduate-0,no-1
# selfemp y-1,n-0
# area semi-1,urban-2,rural-0
```

```python
data['Property_Area'] = le.fit_transform(data['Property_Area'])
data['Gender'] = le.fit_transform(data['Gender'])
data['Loan_Status'] = le.fit_transform(data['Loan_Status'])
data['Education'] = le.fit_transform(data['Education'])
data['Married'] = le.fit_transform(data['Married'])
data['Self_Employed'] = le.fit_transform(data['Self_Employed'])
```

```python
data.head()
```

```python
x = data.drop(columns=['Loan_ID','Loan_Status'],axis=1)
y = data['Loan_Status']
```

```python
x.head()
```

```python
y.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1,stratify=y,random_state=2)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.svm import SVC
from sklearn import svm
```

```python
model = svm.SVC(kernel='linear')
```

```python
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
x_train_pred = model.predict(x_train)
x_train_acc = accuracy_score(y_train,x_train_pred)
x_train_acc
```

```python
x_test_pred = model.predict(x_test)
x_test_acc = accuracy_score(y_test,x_test_pred)
x_test_acc
```

```python
new_data = (0,0,3,1,1,4500,1500.0,200.0,180.0,0.0,0)
new_data_array = np.asarray(new_data)

# Reshape for one instance
new_data_reshaped = new_data_array.reshape(1, -1)

# Standardize the new input (IMPORTANT!)
new_data_scaled = scaler.transform(new_data_reshaped)

# Make prediction
prediction = model.predict(new_data_scaled)

if prediction[0] == 1:
    print('Loan Approved ✅')
else:
    print('Loan Rejected ❌')
```
